```python
import pdfplumber
from pdf2image import convert_from_path
import pytesseract
import google.generativeai as genai

# Configure your Gemini API key
genai.configure(api_key="AIzaSyCwufnQZIaQAo78zYauNHDoK94-7iqhSKE")  # Replace with your actual API key

def extract_text_from_pdf(pdf_file):
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            # Attempt to extract text directly
            extracted_text = page.extract_text()
            if extracted_text:  # If text is found, append it
                text += extracted_text + "\n"
            else:  # If no text found, use OCR
                images = convert_from_path(pdf_file, first_page=page.page_number, last_page=page.page_number)
                for image in images:
                    ocr_text = pytesseract.image_to_string(image)
                    text += ocr_text + "\n"
    return text

def analyze_with_gemini(text):
    model = genai.GenerativeModel("gemini-1.5-flash")
    response = model.generate_content(f"Analyze this newspaper article:\n\n{text}")
    
    # Extracting the generated content safely
    if hasattr(response, 'parts'):
        texts = [part.text for part in response.parts if hasattr(part, 'text')]
        return "\n".join(texts)
    return "No content generated."

# Example usage
pdf_file_path = "TOI Delhi 31-10.pdf"
extracted_text = extract_text_from_pdf(pdf_file_path)

# Analyze or summarize the extracted text using Gemini API
gemini_analysis = analyze_with_gemini(extracted_text)
print(gemini_analysis)
```
